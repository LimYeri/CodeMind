```python
from datasets import load_dataset

dataset_name_summary = "LimYeri/LeetCode_YT_CC_CoT_Summary"
dataset_name_id = "greengerong/leetcode"
dataset_name_content = "kreimben/leetcode_user_submissions"
#Importing the dataset
dataset_summary = load_dataset(dataset_name_summary)
dataset_id = load_dataset(dataset_name_id)
dataset_content = load_dataset(dataset_name_content)
print(dataset_summary)
print(dataset_id)
print(dataset_content)
```

# Data Cleaning

```python
import pandas as pd
df_summary = pd.DataFrame(dataset_summary['train'])
df_id = pd.DataFrame(dataset_id['train'])
df_content = pd.DataFrame(dataset_content['train'])
```

```python
# Data merge
result_df = pd.merge(df_content, df_id, left_on='title_slug', right_on='slug', how='left')
result_df['id'] = result_df['id'].astype('Int64')
```

```python
# Extract only the rows where ['vote_count'] >= 10
result_df = result_df[result_df['vote_count'] >= 10]
result_df = result_df.dropna(subset=['id'])
```

```python
result_df
```

```python
result_df.info()
```

```python
# Remove unnecessary columns.
result_df = result_df.drop(['question_content', 'view_count', 'vote_count', 'slug', 'title', 'difficulty', 'java', 'c++', 'python', 'javascript'], axis=1)

# rename columns
result_df.rename(columns={
    'content_x': 'content',
    'content_y': 'question_content'
}, inplace=True)
```

```python
result_df.info()
```

```python
df_summary.info()
```

```python
# Remove unnecessary columns.
df_summary = df_summary.drop(['cc_content', 'thumbnail', 'java', 'c++', 'python', 'javascript', 'title', 'success_rate', 'total_submission', 'total_accepted', 'question_likes', 'question_dislikes', 'similar_question_ids', 'num_tokens'], axis=1)

# rename columns
df_summary.rename(columns={
    'Summary': 'content'
}, inplace=True)
```

```python
df_summary.info()
```

```python
# Data concat
import pandas as pd
result = pd.concat([df_summary, result_df], ignore_index=True)
```

# Push Dataset to Hub

```python
from datasets import Dataset
dataset = Dataset.from_pandas(result)
dataset
```

```python
dataset.push_to_hub("LimYeri/LeetCode_with_Solutions")
```
